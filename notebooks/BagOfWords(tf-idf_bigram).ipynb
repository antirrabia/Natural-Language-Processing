```python
import pathlib
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn.objects as so
import numpy as np

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.utils import text_dataset_from_directory
```

### Reading the Data

```python
w_dir = pathlib.Path('/home/antirrabia/Documents/DeepLearning-SETs/aclImdb/')
```

```python
tr_d = text_dataset_from_directory( w_dir / 'train' )
va_d = text_dataset_from_directory( w_dir / 'validation' )
te_d = text_dataset_from_directory( w_dir / 'test' )
```

```python
# getting tf-idf bigram
text_vectorization = layers.TextVectorization(
    ngrams=2,
    max_tokens=20000,
    output_mode='tf-idf'
)
```

```python
text_only = tr_d.map( lambda x, y: x )

text_vectorization.adapt( text_only )
```

```python
tr_d_2grams = tr_d.map( lambda x, y: ( text_vectorization(x), y ), num_parallel_calls=tf.data.AUTOTUNE)
va_d_2grams = va_d.map( lambda x, y: ( text_vectorization(x), y ), num_parallel_calls=tf.data.AUTOTUNE)
te_d_2grams = te_d.map( lambda x, y: ( text_vectorization(x), y ), num_parallel_calls=tf.data.AUTOTUNE)
```

```python
tr_d_2grams = tr_d_2grams.prefetch(buffer_size=tf.data.AUTOTUNE)
va_d_2grams = va_d_2grams.prefetch(buffer_size=tf.data.AUTOTUNE)
te_d_2grams = te_d_2grams.prefetch(buffer_size=tf.data.AUTOTUNE)
```

### The Model

```python
strategy = tf.distribute.MirroredStrategy()
```

```python
with strategy.scope():
    inputs = keras.Input(shape=(20000,))
    
    x = layers.Dense(32, activation='relu')(inputs)
    x = layers.Dropout(0.5)(x)
    
    outputs = layers.Dense(1, activation='sigmoid')(x)
    
    model = keras.Model(inputs, outputs)
    
    model.compile(
        optimizer='rmsprop',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
```

```python
callbacks = [
    keras.callbacks.ModelCheckpoint(
        'bigram.keras',
        save_best_only=True,
        monitor='accuracy'
    ),
    keras.callbacks.TensorBoard(
        log_dir='TensorBoard_Logs/'
    )
]

history = model.fit(
    tr_d_2grams.cache(),
    epochs=43,
    validation_data=va_d_2grams.cache(),
    callbacks=callbacks
)
```

### Testing

```python
test_model = keras.models.load_model('bigram.keras')
```

```python
print(f'Test Accuracy: {test_model.evaluate(te_d_2grams)[1]:.3f}')
```

### Learning Curves

```python
accuracy = history.history['accuracy']
val_accu = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range( 1, len(accuracy) + 1 )

fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(25,7))

ax[0].plot(epochs,accuracy, label='Training Accuracy')
ax[0].plot(epochs,val_accu, label='Validation Accuracy')
ax[0].set_title('Training and Validation Accuracy')
ax[0].legend()

ax[1].plot(epochs,loss, label='Training Loss')
ax[1].plot(epochs,val_loss, label='Validation Loss')
ax[1].set_title('Traing and Validation Loss')
ax[1].legend()
```
